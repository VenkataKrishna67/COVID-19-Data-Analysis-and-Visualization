Task 3.2: Find the maximum infection rate for India and China


```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
# Convert problematic columns to numeric, coercing errors to NaN
for column in corona_dataset_aggregated.columns:
    corona_dataset_aggregated[column] = pd.to_numeric(corona_dataset_aggregated[column], errors='coerce')

# Now, try the diff and max operations again
corona_dataset_aggregated.loc["China"].diff().max()
```

Task 3.3: Maximum number of infection rate for all the countries


```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for i in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[i].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
corona_data.head()
```

```python
happiness_report_csv = pd.read_csv("/content/worldwide_happiness_report.csv")
happiness_report_csv.head()
```

```python
#Overall rank vs Healthy life expectancy

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Overall rank']
  ys = series['Healthy life expectancy']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = happiness_report_csv.sort_values('Overall rank', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Overall rank')
_ = plt.ylabel('Healthy life expectancy')
```

```python
#Overall rank vs GDP per capita

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Overall rank']
  ys = series['GDP per capita']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = happiness_report_csv.sort_values('Overall rank', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Overall rank')
_ = plt.ylabel('GDP per capita')
```
